## Step 1: Create Vectors

```python
# TODO: Import numpy
import numpy as np


# TODO: Create two vectors v1 and v2
v1 = np.array([2, 3])
v2 = np.array([1, -1])
```

## Step 2: Vector Addition & Subtraction

```python
# TODO: Add v1 and v2
add = v1 + v2

# TODO: Subtract v2 from v1
sub = v1 - v2

print('add', add)
print('sub', sub)
```

## Step 3: Scalar Multiplication

```python
# TODO: Multiply vector v1 by a scalar (e.g., 3)
mul = 3 * v1

print('mul', mul)
```

## Step 4: Dot Product

```python
# TODO: Compute dot product of v1 and v2
# Dot product = compressed information of element-wise multiplication
#Dot product measures similarity between two vectors (higher = more similar)

#It combines element-wise multiplication into a single meaningful scalar value

#It is the core of neural network computations (weighted sums)

#It powers key AI systems like cosine similarity, embeddings, and attention mechanisms
#dot1 = v1 * v2  -- element-wise multiplication not dot, result [2, -3]
dot2 = np.dot(v1, v2)
dot3 = v1 @ v2
#print('dot1=', dot1) -- element-wise multiplication not dot, result [2, -3]
print('dot2=', dot2)
print('dot3=', dot3)
```

## Step 5: Magnitude (Norm)

```python
# TODO: Find magnitude of v1
#Magnitude represents the scale of features, which can influence model predictions

#It affects distance calculations in algorithms like KNN and clustering

#It impacts similarity measures (e.g., dot product vs cosine similarity)

#Large magnitudes can bias or destabilize models, so normalization is often required
# magnitude/norm  = lenght of the arrow = square root of sum of x and y
v1 = [3, 4]
norm = np.linalg.norm(v1)
print('magnitude=',norm)
```

## Step 6: Normalize Vector

```python
# TODO: Normalize v1
#Removes scale differences

#Makes vectors comparable

#Required for cosine similarity

#Used in embeddings, NLP, recommendation systems
v1 = [3, 4]
norm = np.linalg.norm(v1)
if norm != 0:
    v1_normalized = v1 / norm
print('Normalized v1', v1_normalized)
```

## Step 7: Angle Between Vectors

```python
# TODO: Compute angle between v1 and v2 (in degrees)
# cosine similarity cos theta

#Here are 4 crisp points on cos θ (cosine of angle between vectors):

#Measures similarity in direction → closer to 1 means vectors are more aligned

#Independent of magnitude → focuses only on direction, not size

#Computed using dot product and norms → combines both concepts

#Foundation of cosine similarity → widely used in NLP, search, and recommendation systems

v1 = np.array([2, 3])
v2 = np.array([1, -1])

dot = np.dot(v1, v2)
norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)

cos_theta = dot / (norm_v1 * norm_v2)

#Here are 4 clear points on angle between vectors:

#The angle represents how similar or different the directions of two vectors are

#Smaller angle (close to 0°) means high similarity, larger angle means less similarity

#Angle is derived from dot product and magnitudes of vectors

#It is used in AI via cosine similarity to compare embeddings and features
angle = np.arccos(cos_theta)
angle_degrees = np.degrees(angle)

print("cos_theta:", cos_theta)
print("Angle (degrees):", angle_degrees)
```

## Step 8: Visualization (Optional but Recommended)

## 🚀 Bonus Challenge
- Create 3 vectors
- Find which two are most similar (using dot product)

👉 This is how recommendation systems work!


```python
v1 = np.array([1, 2])
v2 = np.array([2, 4])
v3 = np.array([-1, 1])

sim_12 = np.dot(v1, v2)
sim_13 = np.dot(v1, v3)
sim_23 = np.dot(v2, v3)

print('Dot Similarity=',sim_12, sim_13, sim_23)
similarities = {
    "v1 & v2": sim_12,
    "v1 & v3": sim_13,
    "v2 & v3": sim_23
}

most_similar = max(similarities, key=similarities.get)

print("Most similar pair:", most_similar)
```

```python
# Dot product depends on magnitude also 
# Cosine Similarity — REAL AI WAY
v1 = np.array([1, 2])
v2 = np.array([2, 4])
v3 = np.array([-1, 1])

from numpy.linalg import norm

def cosine_sim(a, b):
    return np.dot(a, b) / (norm(a) * norm(b))

sim_12 = cosine_sim(v1, v2)
sim_13 = cosine_sim(v1, v3)
sim_23 = cosine_sim(v2, v3)
print('Cosine Similarity=',sim_12, sim_13, sim_23)

similarities = {
    "v1 & v2": sim_12,
    "v1 & v3": sim_13,
    "v2 & v3": sim_23
}

most_similar = max(similarities, key=similarities.get)

print("Most similar pair:", most_similar)
```
