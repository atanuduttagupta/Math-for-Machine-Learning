```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We will use numpy for calculations
# matplotlib for visualization

# ============================================================
# 2. CREATE SAMPLE DATA
# ============================================================
```

```python
# Let's start with a simple dataset (marks of students)
data = np.array([10, 20, 30, 40, 50])
# This is a "clean" dataset (no outliers)
print("Dataset:", data)
```

```python
# ============================================================
# 3. MEAN (AVERAGE)
# ============================================================
# Mean = sum of all values / total number of values
mean_value = np.mean(data)
print("Mean: ", mean_value)

# 👉 Intuition:
# Mean is like the "center of gravity" of data
# If all values were placed on a number line, mean is the balancing point
```

```python
# ============================================================
# 4. MEDIAN
# ============================================================
# Median = middle value after sorting
median_value = np.median(data)
print("Median: ", median_value)
# 👉 Intuition:
# Median is the "middle person" in sorted data
# It is NOT affected much by extreme values (outliers)
```

```python
# ============================================================
# 5. MODE (MANUAL)
# ============================================================

# Numpy doesn't directly give mode easily, so let's create one example
from collections import Counter
mode_data = [1,2,2,3,4]
mode_value = Counter(mode_data).most_common(1)[0][0]
print(mode_value)
```

```python
# 👉 Intuition:
# Mode = most frequent value
# Useful in categorical data (e.g., most common product)
```

```python
# ============================================================
# 6. ADD AN OUTLIER (VERY IMPORTANT)
# ============================================================

# Let's introduce an extreme value
outlier_data = np.array([10,20,30,40,100])
print("\nDataset with outlier:", outlier_data)

print("Mean with outlier:", np.mean(outlier_data))
print("Median with outlier:", np.median(outlier_data))
```

```python
# 👉 Observe:
# Mean shifts a LOT because of 100
# Median stays relatively stable

# 👉 KEY LEARNING:
# Mean is sensitive to outliers
# Median is robust
```

```python
# ============================================================
# 7. VARIANCE
# ============================================================
variance_val = np.var(data)
print("\nVariance: ", variance_val)
```

```python
# 👉 Intuition:
# Variance measures how far data points are from the mean
# Larger variance = more spread out data

# Why squared?
# Because differences can be negative
# Squaring makes everything positive + penalizes large deviations more
```

```python
# ============================================================
# 8. STANDARD DEVIATION
# ============================================================
std_dev = np.std(data)
print('std_dev: ', std_dev)
```

```python
# 👉 Intuition:
# Standard deviation = square root of variance
# Easier to interpret because it's in SAME units as data
```

```python
# Let's plot histogram
plt.hist(outlier_data, bins=5)
plt.title("Data Distribution (with outlier)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

# 👉 Intuition:
# Histogram shows how data is distributed
# You can SEE skew caused by outlier
```

```python
# ============================================================
# 10. MINI PROJECT: STUDENT ANALYSIS
# ============================================================

marks = np.array([45, 50, 55, 60, 65, 90])

```

```python
# Task 1: Basic statistics
print("Mean:", np.mean(marks))
print("Median:", np.median(marks))
print("Std Dev:", np.std(marks))
```

```python
# Task 2: Detect possible outlier
# Simple rule: value far from mean
mean_marks = np.mean(marks)
std_marks = np.std(marks)


```

```python
# Let's flag anything beyond 2 standard deviations
outliers = marks[np.abs(marks - mean_marks) > 2 * std_marks]
print("Possible Outliers:", outliers)
```

```python
# Task 3: Visualization
plt.hist(marks, bins=6)
plt.title("Student Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```
