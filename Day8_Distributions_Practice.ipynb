# 🔹 Section 1: Setup

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```


# 🔹 Section 2: Task 1 - Bernoulli Distribution


```python
## Simulate click (1) or no click (0)

p = 0.3
samples = np.random.binomial(n=1, p=p, size = 1000)

print("Bernoulli Mean (Expected CTR):", np.mean(samples))

plt.hist(samples, bins=10)
plt.title("Bernoulli Distribution (Click / No Click)")
plt.show()
```

# 🔹 Section 3: Task 2 - Binomial Distribution

```python
# Number of clicks out of 10 users

n= 10
p = .3

samples = np.random.binomial(n=n, p=p, size = 1000)
print('binomial means=', np.mean(samples))

plt.hist(samples, bins=10)
plt.title("Binomial Distribution (Clicks out of 10 users)")
plt.show()
```

# 🔹 Section 4: Task 3 - Poisson Distribution

```python
# Events per interval (e.g., clicks per minute)


lambda_val = 5
samples = np.random.poisson(lam=lambda_val, size = 1000)

print("Poisson Mean:", np.mean(samples))

plt.hist(samples, bins=15)
plt.title("Poisson Distribution (Events per interval)")
plt.show()
```

# 🔹 Section 5: Task 4 - Normal Distribution

```python
# Session duration (in minutes)
mu = 10
sigma = 2
samples = np.random.normal(loc=mu, scale=sigma, size=1000)

print("Normal Mean:", np.mean(samples))
print("Normal Std:", np.std(samples))

plt.hist(samples, bins=20)
plt.title("Normal Distribution (Session Duration)")
plt.show()
```

# 🔹 Section 6: Task 5 - Uniform Distribution

```python
samples = np.random.uniform(low=0, high=1, size=1000)

print("Uniform Mean:", np.mean(samples))

plt.hist(samples, bins=20)
plt.title("Uniform Distribution")
plt.show()
```

# Exponential Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Average rate = 5 events per minute
lambda_val = 5

# Time between events = exponential
samples = np.random.exponential(scale=1/lambda_val, size=1000)

print("Mean waiting time:", np.mean(samples))

plt.hist(samples, bins=20)
plt.title("Exponential Distribution (Time Between Clicks)")
plt.show()
```

# 🔹 Section 7: Mini Project - CTR Simulation (Advanced)

```python
# Simulating multiple user segments
users = 1000

segments = {
    "casual": {"ctr": 0.02},
    "regular": {"ctr": 0.05},
    "power": {"ctr": 0.1}
}

results = {}

for seg, vals in segments.items():
    clicks = np.random.binomial(n=1, p=vals["ctr"], size=users)
    results[seg] = clicks

# Plot comparison
for seg in results:
    plt.hist(results[seg], alpha=0.5, label=seg)

plt.legend()
plt.title("CTR Distribution Across User Segments")
plt.show()

```

# 🔹 Section 8: Insights

```python
# Calculate CTR per segment
for seg in results:
    ctr = np.mean(results[seg])
    print(f"{seg} CTR: {ctr:.4f}")

# Overall CTR
all_clicks = np.concatenate(list(results.values()))
print("Overall CTR:", np.mean(all_clicks))
```
