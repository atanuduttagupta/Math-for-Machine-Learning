## 🧠 Section 6: PDF (Continuous Case)

Time spent on a website follows a continuous distribution

```python
# Task 7:
# Explain why PDF is used instead of PMF

# YOUR SOLUTION:
# PDF → probability of interval (area under curve)
# PDF is defined as a function (not dictionary)

def cdf(x):
    if x < 0:
        return 0
    elif 0 <= x <= 10:
        return x / 10
    else:
        return 1

print(cdf(6))
```
