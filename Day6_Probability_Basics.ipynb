## 🧠 Task 2: Define Events

```python
# Event A: Even numbers
# Event B: Numbers greater than 4

# TODO: Define A
A = {2,4,6}

# TODO: Define B
B = {5,6}
```

## 🧠 Task 3: Basic Probability

```python
# TODO: Compute P(A)
#p = favorable outcome/total outcome

total = len(S)
P_A = len(A)/total
P_B = len(B)/total

print(f'P_A={P_A}')
# TODO: Compute P(B)
print(f'P_B={P_B}')
```

## 🧠 Task 4: Union of Events

```python
# TODO: Compute A ∪ B
A_U_B = A | B
P_A_U_B = len(A|B)/total
print('P_A_U_B=', P_A_U_B)
# TODO: Compute P(A ∪ B)
```

## 🧠 Task 5: Intersection of Events

```python
# TODO: Compute A ∩ B
A_int_B = A & B
P_A_int_B = len(A & B)/total
print('A_int_B=', A_int_B)
print('P_A_int_B=', P_A_int_B)
# TODO: Compute P(A ∩ B)
```

## 🧠 Task 6: Conditional Probability

```python
# Formula: P(A|B) = P(A ∩ B) / P(B)
A_int_B = A & B
P_A_int_B = len(A_int_B)/total
print('P_A_int_B=', P_A_int_B)

P_B = len(B)/total

P_A_given_b = P_A_int_B / P_B
print('P_A_given_b=', P_A_given_b)
# TODO: Compute P(A | B)
```

## 🧠 Task 7: Independent Events (Coin Toss)

```python
# Sample space: HH, HT, TH, TT

# TODO: Define sample space
S = {'HH', 'TT', 'HT', 'TH'}
# TODO: Event C: First coin is Head
C = {'HH', 'HT'}
# TODO: Event D: Second coin is Head
D = {'HH', 'TH'}
# TODO: Check if independent:
P_C = len(C)/len(S)
P_D = len(D)/len(S)
P_C_int_D = len(C & D)/len(S)
# P(C ∩ D) == P(C) * P(D)
print("P(C):", P_C)
print("P(D):", P_D)
print("P(C ∩ D):", P_C_int_D)
print("P(C) * P(D):", P_C * P_D)
print("Are C and D independent?", P_C_int_D == P_C * P_D)
```

## 🚀 Mini Project: Ad Click Probability Simulator

🎯 Goal: Estimate probability of users clicking an ad

**Scenario:**
- 1000 users visit a website
- Some click ads, some don’t

👉 You will simulate and compute probabilities


```python
# Step 1: Simulate user clicks
# TODO: Create random 0/1 data (1 = clicked)
import numpy as np
np.random.seed(42)

clicks = np.random.choice([0,1], size = 1000, p=(.7, .3))
print(clicks[:20])
# Step 2: Compute probability of click
# TODO: P(click)
p_clicks = np.sum(clicks)/len(clicks)
print(f'p_clicks = {p_clicks}')
# Step 3: Conditional probability
# Example: Given mobile user, probability of click
user_types = np.random.choice([0,1], size = 1000, p = (.5,.5))
mobile_user = (user_types==1)
mobile_click = clicks[mobile_user]
p_click_given_mobile = np.sum(mobile_click)/len(mobile_click)#p_click_and_mobile/p_mobile
print(f'p_click_given_mobile = {p_click_given_mobile}')
# TODO: Simulate user type and compute conditional probability
```
