## 📅 Day 07 - Bayes Theorem (Medical Diagnosis Example - Detailed Version)

### ============================================================
### 🎯 LEARNING GOAL
### ============================================================
### In this notebook, you will:
### 1. Understand Bayes Theorem using a REAL medical scenario
### 2. Learn how probability updates after new evidence
### 3. Build strong intuition (VERY IMPORTANT for AI/ML)
###
### Key Question:
### 👉 If a test is 99% accurate and you test positive,
###    do you REALLY have the disease?
###
### Most people think "YES" — but Bayes will surprise you!

```python
# ============================================================
# 🧠 STEP 0: DEFINE THE PROBLEM
# ============================================================
# Suppose:
# - Only 1% of people have the disease (rare disease)
# - Test is 99% accurate
#
# We define probabilities:

# P(Disease) → Prior probability (before test)
P_disease = 0.01

# P(No Disease)
P_no_disease = 1 - P_disease

# Likelihoods:
# P(Positive | Disease) → True Positive Rate
P_pos_given_disease = 0.99

# P(Positive | No Disease) → False Positive Rate
P_pos_given_no_disease = 0.01
```

```python
# ============================================================
# 🧮 STEP 1: TOTAL PROBABILITY OF TESTING POSITIVE
# ============================================================
# IMPORTANT CONCEPT:
# A positive test can come from TWO cases:
#   1. Person has disease AND test detects it
#   2. Person does NOT have disease BUT test is wrong
#
# So we calculate:
# P(Positive) =
#   (Positive from Disease) + (Positive from No Disease)

P_positive = (P_pos_given_disease * P_disease) + \
             (P_pos_given_no_disease * P_no_disease)

print("Step 1: Total probability of testing positive:")
print("P(Positive) =", round(P_positive, 4))
```

```python
# ============================================================
# 🧮 STEP 2: APPLY BAYES THEOREM
# ============================================================
# Bayes Formula:
# P(Disease | Positive) =
#   (P(Positive | Disease) * P(Disease)) / P(Positive)
#
# Interpretation:
# 👉 Out of ALL positive tests, how many are TRUE disease cases?
```

```python
P_disease_given_positive = (P_pos_given_disease * P_disease) / P_positive

print("\nStep 2: Updated probability after test:")
print("P(Disease | Positive) =", round(P_disease_given_positive, 4))
```

```python
# ============================================================
# 😮 STEP 3: INTERPRETATION (MOST IMPORTANT)
# ============================================================
print("\n--- INTERPRETATION ---")

if P_disease_given_positive < 0.5:
    print("Even after testing positive, probability is LESS than 50%!")
    print("Reason: Disease is very rare → many false positives")
else:
    print("High probability of actually having the disease.")
```

```python
# ============================================================
# 📊 STEP 4: INTUITION WITH 1000 PEOPLE (SIMULATION STYLE)
# ============================================================
# This section helps you THINK in real numbers instead of abstract probability

print("\n--- INTUITION USING 1000 PEOPLE ---")

population = 1000

# People with disease
disease_people = population * P_disease

# People without disease
healthy_people = population * P_no_disease

# True positives (correct detection)
true_positives = disease_people * P_pos_given_disease

# False positives (wrong detection)
false_positives = healthy_people * P_pos_given_no_disease

print(f"Out of {population} people:")
print(f"- Disease cases: {int(disease_people)}")
print(f"- True positives: {int(true_positives)}")
print(f"- False positives: {int(false_positives)}")

print("\n👉 Total positive tests = True + False")
total_positive = true_positives + false_positives
print("Total positives:", int(total_positive))

print("👉 Actual probability:")
print("= True positives / Total positives")
print("=", round(true_positives / total_positive, 4))
```

```python
# ============================================================
# 🧪 STEP 5: EXPERIMENT - CHANGE DISEASE RATE
# ============================================================
# KEY INSIGHT:
# Bayes heavily depends on PRIOR (P(Disease))

print("\n--- EXPERIMENT: Changing Disease Prevalence ---")

for prevalence in [0.01, 0.05, 0.1, 0.2]:
    P_disease = prevalence
    P_no_disease = 1 - prevalence

    P_positive = (P_pos_given_disease * P_disease) + \
                 (P_pos_given_no_disease * P_no_disease)

    P_disease_given_positive = (P_pos_given_disease * P_disease) / P_positive

    print(f"Disease Rate: {prevalence*100:.0f}% -> Posterior: {round(P_disease_given_positive, 3)}")
```
