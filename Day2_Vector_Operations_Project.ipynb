# 📘 Day 2: Vector Operations Project

## 🎯 Objective
Build deeper understanding of vector operations and apply them to a mini AI-style similarity system.



## Step 1: Create Multiple Vectors

```python
# TODO: Import numpy
import numpy as np
# TODO: Create at least 4 vectors representing items e.g. movies
# user preference of movies
# 3 features
# [Thrill, Action, Comedy]
movie1 = np.array([1, 0, 0])   # (1,0)
movie2 = np.array([0, 1, 0])   # (0,1)
movie3 = np.array([1, 1, 0])   # (1,1)
movie4 = np.array([2, 1, 1])   # (2,1)
```

## Step 2: Element-wise Operations

```python
# TODO: Perform element-wise multiplication
#Multiplication	Common features
#Addition	Combined features
#Subtraction	Difference between items
mul = movie1 * movie2
# TODO: Perform addition and subtraction
add = movie1 + movie2
sub = movie1 - movie2

print("Multiplication:", mul)
print("Addition:", add)
print("Subtraction:", sub)
```

## Step 3: Dot Product Similarity

```python
# TODO: Compute dot product between vector pairs
#Dot product = count of shared features (in this case)
sim_12 = np.dot(movie1, movie2)
sim_13 = np.dot(movie1, movie3)
sim_14 = np.dot(movie1, movie4)
sim_23 = np.dot(movie2, movie3)
sim_24 = np.dot(movie2, movie4)
sim_34 = np.dot(movie3, movie4)

print('sim_12 =',sim_12)
print('sim_13 =',sim_13)
print('sim_14 =',sim_14)
print('sim_23 =',sim_23)
print('sim_24 =',sim_24)
print('sim_34 =',sim_34)
```

## Step 4: Magnitude & Normalization

```python
#Normalization means:“Make THIS vector length = 1”
# TODO: Compute magnitude(norm)
movie1_magn = np.linalg.norm(movie1)
movie2_magn = np.linalg.norm(movie2)
movie3_magn = np.linalg.norm(movie3)
movie4_magn = np.linalg.norm(movie4)
# TODO: Normalize vectors
movie1_norm = movie1/movie1_magn
movie2_norm = movie2/movie2_magn
movie3_norm = movie3/movie3_magn
movie4_norm = movie4/movie4_magn

print('movie1_norm =',movie1_norm)
print('movie2_norm =',movie2_norm)
print('movie3_norm =',movie3_norm)
print('movie4_norm =',movie4_norm)
```

## Step 5: Cosine Similarity Function

```python
# TODO: Compare all vectors
from numpy.linalg import norm
# TODO: Store results
def cosine_similarity(v1, v2):
    return np.dot(v1,v2)/(norm(v1)*norm(v2))
# TODO: Find most similar pair
print("movie1 vs movie2:", cosine_similarity(movie1, movie2))
print("movie1 vs movie3:", cosine_similarity(movie1, movie3))
print("movie1 vs movie4:", cosine_similarity(movie1, movie4))
```

## Step 6: Mini Similarity Engine

```python
# TODO: Compare all vector pairs
sim12 = cosine_similarity(movie1, movie2)
sim13 = cosine_similarity(movie1, movie3)
sim14 = cosine_similarity(movie1, movie4)
sim23 = cosine_similarity(movie2, movie3)
sim24 = cosine_similarity(movie2, movie4)
sim34 = cosine_similarity(movie3, movie4)
# TODO: Store similarities in a dictionary
similarities = {
    'movie1-movie2': sim12,
    'movie1-movie3': sim13,
    'movie1-movie4': sim14,
    'movie2-movie3': sim23,
    'movie2-movie4': sim24,
    'movie3-movie4': sim34,
}
# TODO: Find most similar pair
most_similar = max(similarities, key=similarities.get)
print('Most Similar two Movies=', most_similar)
print('Similarity=', similarities[most_similar])
```
